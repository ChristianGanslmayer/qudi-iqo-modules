```python
import sys
import os
import subprocess
from pathlib import Path
import time
import datetime
import numpy as np
from scipy import optimize
from qudi.util.network import netobtain
import matplotlib.pyplot as plt
import json
import importlib
sys.path.append('../xq1i')
import libxq1i
importlib.reload(libxq1i)

confocal2 = libxq1i.xq1i(pulsed_master_logic, pulsed_measurement_logic, sequence_generator_logic)
```

```python
print("Started at:", datetime.datetime.now().strftime("%H:%M"))
# ------------------------------------------------------------------
# 1) SELECT WHICH MEASUREMENTS TO PERFORM + NUMBER OF SWEEPS
# ------------------------------------------------------------------

confocal2.generate_params['microwave_amplitude'] = 100e-3

selected_routines = {
    "Rabi": {
        "enabled": True,
        "sweeps": 5e5
    },
    "ODMR": {
        "enabled": True,
        "sweeps": 5e5
    },
    "DEERsim": {
        "enabled": True,
        "sweeps": 5e5
    }
}

# ------------------------------------------------------------------
# 2) SETUP: Folder Creation, Orientation Props, POI List
# ------------------------------------------------------------------
results_root = r"C:\Users\XQ1\qudi\Data\2025\07\DEER_auto"
timestamp_str = datetime.datetime.now().strftime('%Y%m%d_%H%M')
run_folder_name = f"_{timestamp_str}"
run_folder = os.path.join(results_root, run_folder_name)
os.makedirs(run_folder, exist_ok=True)

# Dictionary to store all measurement results in tabulate-friendly form
all_results = []

orientation_properties = {
    1: {'res_freq': 2.639e9, 'freq_range': [2.63e9 - 1e6 * 25, 1e6, 50], 'rabi_period': 60e-9},
    2: {'res_freq': 2.792e9, 'freq_range': [2.79e9 - 800e3 * 25, 800e3, 50], 'rabi_period': 382e-9},
    3: {'res_freq': 2.827e9, 'freq_range': [2.83e9 - 800e3 * 25, 800e3, 50], 'rabi_period': 422e-9},
    # 4: ...
}

poi_list = []

# Spot8: POI1 to POI24
#for i in range(13, 25):
#    poi_list.append({'poi': f'Spot8_POI{i}', 'orientations': [1, 2, 3]})

# Spot6: POI25 to POI42
for i in range(25, 43):
    poi_list.append({'poi': f'Spot6_POI{i}', 'orientations': [1, 2, 3]})

# Clear pulse generator
pulsed_master_logic.clear_pulse_generator()
time.sleep(1)


# ------------------------------------------------------------------
# 3) HELPER: Safe formatting (None -> "N/A")
# ------------------------------------------------------------------
def fmt_or_na(val, fmt="{:.3E}"):
    """Format numerical value or return 'N/A' if None or not a number."""
    if val is None:
        return "N/A"
    try:
        return fmt.format(val)
    except:
        return str(val)

    
    
    
# -----------------------------------------------------------
# 4) RABI - Function
# -----------------------------------------------------------
def perform_rabi_measurement(
    poi, 
    orientation, 
    microwave_frequency, 
    rabi_sweeps, 
    run_folder, 
    label_suffix="Rabi_1"
):
    """
    Performs a Rabi measurement, fits the data, and saves the results.

    Args:
        poi (str): The current Point of Interest.
        orientation (int): The current orientation number.
        microwave_frequency (float): The microwave frequency to use for the measurement.
        rabi_sweeps (int): The number of sweeps for the measurement.
        run_folder (str): The path to the folder where results should be saved.
        label_suffix (str): A suffix for the filename and plot title.

    Returns:
        A dictionary containing the extracted fit parameters and fit quality metrics.
    """
    
    # Initialize variables for results
    rabi_period_val, rabi_period_err = None, None
    rabi_contrast_val, rabi_contrast_err = None, None
    rabi_rsquared, rabi_redchi = None, None

    print(f"Rabi measurement ({label_suffix}) started at:", datetime.datetime.now().strftime("%H:%M"))
    
    # Configure and run the measurement
    confocal2.generate_params['microwave_frequency'] = microwave_frequency
    pulsed_master_logic.set_generation_parameters(confocal2.generate_params)
    confocal2.rabi_sweeps = rabi_sweeps

    rabi_label = f"{poi}_ornt_{orientation}_{label_suffix}"
    print(f"Performing Rabi measurement: {rabi_label} with {confocal2.rabi_sweeps} sweeps")
    confocal2.do_rabi(rabi_label, isSlow=True)
    
    try:
        # Fit and extract parameters
        result_rabi = netobtain(pulsed_measurement_logic.do_fit('Sine'))
        amp_val = result_rabi.params['amplitude'].value
        amp_err = result_rabi.params['amplitude'].stderr
        off_val = result_rabi.params['offset'].value
        off_err = result_rabi.params['offset'].stderr
        rabi_period_val = result_rabi.params['period'].value
        rabi_period_err = result_rabi.params['period'].stderr
        rabi_rsquared = result_rabi.rsquared
        rabi_redchi = result_rabi.redchi
        
        # Calculate contrast
        if off_val and amp_val is not None:
            rabi_contrast_val = amp_val / off_val
            if all([amp_err, off_err, amp_val != 0, off_val != 0]):
                rel_err_sq = (amp_err / amp_val)**2 + (off_err / off_val)**2
                rabi_contrast_err = abs(rabi_contrast_val) * np.sqrt(rel_err_sq)

        # Save raw Rabi data
        tData = netobtain(pulsed_measurement_logic.signal_data[0])
        sigData = netobtain(pulsed_measurement_logic.signal_data[1])
        sigData_err = netobtain(pulsed_measurement_logic.measurement_error[1])
        rabi_data_filename = os.path.join(run_folder, f"{rabi_label}.txt")
        with open(rabi_data_filename, 'w') as f:
            f.write("# time (s)\tcounts\terr\n")
            for x, y, z in zip(tData, sigData, sigData_err):
                f.write(f"{x}\t{y}\t{z}\n")

        # Plot & save Rabi figure
        plt.figure(figsize=(7, 1.5))
        plt.plot(tData, sigData, 'o-', ms=3, linewidth=0.5, label='data')
        plt.errorbar(tData, sigData, sigData_err, fmt='o', ms=3, linewidth=0.5, ecolor='red')
        plt.plot(result_rabi.high_res_best_fit[0], result_rabi.high_res_best_fit[1], label='fit')
        plt.title(f"Rabi: T={fmt_or_na(rabi_period_val)} s")
        plt.xlabel("time (s)")
        plt.ylabel("counts")
        plt.grid()
        plt.legend()
        rabi_fig_path = os.path.join(run_folder, f"{rabi_label}.png")
        plt.savefig(rabi_fig_path, dpi=150, bbox_inches='tight')
        plt.show()

    except Exception as e:
        print(f"Error during Rabi measurement or fit for {rabi_label}: {e}")

    return {
        "period_val": rabi_period_val,
        "period_err": rabi_period_err,
        "contrast_val": rabi_contrast_val,
        "contrast_err": rabi_contrast_err,
        "rsquared": rabi_rsquared,
        "redchi": rabi_redchi
    }
# ------------------------------------------------------------------
# 4) MAIN LOOP OVER POIs AND ORIENTATIONS
# ------------------------------------------------------------------



for entry in poi_list:
    currPoi = entry['poi']
    if len(entry['orientations']) < 2:
        print(f"Skipping {currPoi}: only 1 orientation")
        continue  # Skip POIs with less than 2 visible orientations 
    
    num_orientations = len(entry['orientations'])
    odmr_frequencies = [None] * num_orientations  
    rabi_periods = [None] * num_orientations
    odmr_rsquares = [None] * num_orientations  
    rabi_rsquares = [None] * num_orientations  
    
    print(f"Go to POI {currPoi} ...")

    # Disable refocus/tracking, move to POI, then re-enable
    poi_manager_logic.toggle_periodic_refocus(False)
    while (
        scanner_gui.module_state() == 'locked'
        or (scanning_optimize_logic._sequence_index < len(scanning_optimize_logic._scan_sequence))
    ):
        time.sleep(0.1)

    poi_manager_logic.go_to_poi(currPoi)
    poi_manager_logic.set_active_poi(currPoi)
    poi_manager_logic.toggle_periodic_refocus(True)
    while (
        scanner_gui.module_state() == 'locked'
        or (scanning_optimize_logic._sequence_index < len(scanning_optimize_logic._scan_sequence))
    ):
        time.sleep(0.1)

    print("Position and counts at peak after refocus:", scanner_gui.optimizer_dockwidget.result_label.text())
    for orientation in entry['orientations']:

        orientation_start_time = time.time()  # <--- Start timing here
        
        print(f"Start measurements for orientation {orientation} ...")
        countsStart = scanner_gui.optimizer_dockwidget.result_label.text()[-9:]

        # Grab orientation properties
        props = orientation_properties[orientation]
        res_freq = props['res_freq']

        # Initialize placeholders for all measurement results
        rabi_period_val     = None
        rabi_period_err     = None
        rabi_contrast_val   = None
        rabi_contrast_err   = None
        rabi_rsquared       = None
        rabi_redchi         = None

        odmr_center_val     = None
        odmr_center_err     = None
        odmr_sigma_val      = None
        odmr_sigma_err      = None
        odmr_contrast_val   = None
        odmr_contrast_err   = None
        odmr_rsquared       = None
        odmr_redchi         = None

        rabi_period_val_2     = None
        rabi_period_err_2     = None
        rabi_contrast_val_2   = None
        rabi_contrast_err_2   = None
        rabi_rsquared_2       = None
        rabi_redchi_2         = None

        deer_period_val     = None
        deer_period_err     = None
        deer_contrast_val   = None
        deer_contrast_err   = None
        deer_rsquared       = None
        deer_redchi         = None

        hahn_t2_val         = None
        hahn_t2_err         = None
        hahn_rsquared       = None
        hahn_redchi         = None



        # -----------------------------------------------------------
        # 4A) RABI
        # -----------------------------------------------------------
        if selected_routines["Rabi"]["enabled"]:
            rabi_results_1 = perform_rabi_measurement(
                currPoi, 
                orientation, 
                res_freq, 
                selected_routines["Rabi"]["sweeps"], 
                run_folder,
                label_suffix="Rabi_1"
            )
            rabi_period_val = rabi_results_1["period_val"]
            rabi_period_err = rabi_results_1["period_err"]
            rabi_contrast_val = rabi_results_1["contrast_val"]
            rabi_contrast_err = rabi_results_1["contrast_err"]
            rabi_rsquared = rabi_results_1["rsquared"]
            rabi_redchi = rabi_results_1["redchi"]
            
            # Store the period and R-squared for later comparison
            rabi_periods[int(orientation)-1] = float(rabi_period_val) if rabi_period_val is not None else None
            rabi_rsquares[int(orientation)-1] = float(rabi_rsquared) if rabi_rsquared is not None else None
        else:
            print("Skipping Rabi measurement.")
            # Use fallback period from orientation properties
            rabi_period_val = props['rabi_period']

        # -----------------------------------------------------------
        # 4B) ODMR
        # -----------------------------------------------------------
        print("ODMR started at:", datetime.datetime.now().strftime("%H:%M"))
        if selected_routines["ODMR"]["enabled"]:
            confocal2.pulsedODMR_params['freq_start']    = props['freq_range'][0]
            confocal2.pulsedODMR_params['freq_step']     = props['freq_range'][1]
            confocal2.pulsedODMR_params['num_of_points'] = props['freq_range'][2]

            # Use measured or fallback Rabi period
            confocal2.generate_params['rabi_period'] = rabi_period_val
            pulsed_master_logic.set_generation_parameters(confocal2.generate_params)
            confocal2.pulsedODMR_sweeps = selected_routines["ODMR"]["sweeps"]

            odmr_label = f"{currPoi}_ornt_{orientation}_ODMR"
            print(f"Performing ODMR measurement: {odmr_label} with {confocal2.pulsedODMR_sweeps} sweeps")
            confocal2.do_pulsedODMR(odmr_label)

            result_odmr = netobtain(pulsed_measurement_logic.do_fit('Lorentzian Dip'))
            cen_val = result_odmr.params['center'].value
            cen_err = result_odmr.params['center'].stderr
            amp_val = result_odmr.params['amplitude'].value
            amp_err = result_odmr.params['amplitude'].stderr
            off_val = result_odmr.params['offset'].value
            off_err = result_odmr.params['offset'].stderr

            # Some fits define "sigma", "gamma", or "fwhm"
            if 'sigma' in result_odmr.params:
                sig_val = result_odmr.params['sigma'].value
                sig_err = result_odmr.params['sigma'].stderr
            else:
                sig_val = None
                sig_err = None

            odmr_rsquared = result_odmr.rsquared
            odmr_redchi   = result_odmr.redchi
            odmr_rsquares[int(orientation)-1] = float(odmr_rsquared)
            odmr_center_val = cen_val
            odmr_center_err = cen_err
            odmr_sigma_val  = sig_val
            odmr_sigma_err  = sig_err

            if off_val and amp_val is not None:
                cval = -amp_val / off_val
                if all([amp_err, off_err, amp_val != 0, off_val != 0]):
                    rel_err_sq = (amp_err / amp_val)**2 + (off_err / off_val)**2
                    cerr = abs(cval) * np.sqrt(rel_err_sq)
                else:
                    cerr = None
                odmr_contrast_val = cval
                odmr_contrast_err = cerr


            # Grab raw ODMR data
            fData_odmr = netobtain(pulsed_measurement_logic.signal_data[0])
            sigData_odmr = netobtain(pulsed_measurement_logic.signal_data[1])
            sigData_odmr_err = netobtain(pulsed_measurement_logic.measurement_error[1])

            # Save raw ODMR data
            odmr_data_filename = os.path.join(run_folder, f"ODMR_{currPoi}_ornt_{orientation}.txt")
            with open(odmr_data_filename, 'w') as f:
                f.write("# frequency (Hz)\tcounts\terr\n")
                for fx, sy, z in zip(fData_odmr, sigData_odmr, sigData_odmr_err):
                    f.write(f"{fx}\t{sy}\t{z}\n")

            # Plot & save ODMR figure
            plt.figure(figsize=(7, 1.5))
            plt.plot(fData_odmr, sigData_odmr, 'o-', ms=3, linewidth=0.5, label='data')
            plt.errorbar(fData_odmr, sigData_odmr, sigData_odmr_err, fmt='o', ms=3, linewidth=0.5, ecolor='red')
            plt.plot(result_odmr.high_res_best_fit[0], result_odmr.high_res_best_fit[1], label='fit')
            plt.title(f"ODMR: f_res={odmr_center_val:.3E} Hz, contrast={odmr_contrast_val:.2f}")
            plt.xlabel("frequency (Hz)")
            plt.ylabel("counts")
            plt.grid()
            plt.legend()
            odmr_fig_path = os.path.join(run_folder, f"ODMR_{currPoi}_ornt_{orientation}.png")
            plt.savefig(odmr_fig_path, dpi=150, bbox_inches='tight')
            plt.show()

            while (
                scanner_gui.module_state() == 'locked'
                or (scanning_optimize_logic._sequence_index < len(scanning_optimize_logic._scan_sequence))
            ):
                time.sleep(0.1)
            countsEnd = scanner_gui.optimizer_dockwidget.result_label.text()[-9:]
            
            odmr_frequencies[int(orientation)-1] = float(odmr_center_val)
            
            #go to next if ODMR is bad
            if abs(float(odmr_rsquared)) < 0.6:
                print("bad ODMR contrast, aborting measurement")
                continue

        else:
            print("Skipping ODMR measurement.")
            countsEnd = scanner_gui.optimizer_dockwidget.result_label.text()[-9:]

        # -----------------------------------------------------------
        # 4C) Second RABI
        # -----------------------------------------------------------
        if selected_routines["Rabi"]["enabled"]:
            if odmr_center_val is not None:
                rabi_results_2 = perform_rabi_measurement(
                    currPoi, 
                    orientation, 
                    odmr_center_val, 
                    selected_routines["Rabi"]["sweeps"], 
                    run_folder,
                    label_suffix="Rabi_2"
                )
                rabi_period_val_2 = rabi_results_2["period_val"]
                rabi_period_err_2 = rabi_results_2["period_err"]
                rabi_contrast_val_2 = rabi_results_2["contrast_val"]
                rabi_contrast_err_2 = rabi_results_2["contrast_err"]
                rabi_rsquared_2 = rabi_results_2["rsquared"]
                rabi_redchi_2 = rabi_results_2["redchi"]

                # Update the Rabi period if the second fit is better
                if rabi_rsquared_2 is not None and rabi_rsquares[int(orientation)-1] is not None and rabi_rsquares[int(orientation)-1] < rabi_rsquared_2:
                    rabi_periods[int(orientation)-1] = float(rabi_period_val_2)
            else:
                print("Skipping second Rabi measurement: ODMR frequency not found.")
        else:
            print("Skipping Rabi measurement.")
            rabi_period_val = props['rabi_period'] # Use fallback if necessary
  
    # -----------------------------------------------------------
    # 5) DEER
    # -----------------------------------------------------------

    # Loop over pairs of orientations for DEER measurement
    for i in range(len(entry['orientations'])):
        for j in range(len(entry['orientations'])):
            if i == j:
                continue  # Skip DEER between the same orientation

            if (
                odmr_rsquares[i] is None or
                odmr_rsquares[j] is None or
                odmr_rsquares[i] < 0.5 or
                odmr_rsquares[j] < 0.5
            ):
                print(f"Skipping pair (i={i}, j={j}) due to low ODMR confidence: {odmr_rsquares}")
                continue

            print(f"DEER {i+1},{j+1} started at:", datetime.datetime.now().strftime("%H:%M"))

            deer_label = f"{currPoi}_{i+1},{j+1}_DEER"

            # ... (unchanged configuration code for normal DEER) ...
            confocal2.generate_params['microwave_frequency'] = float(odmr_frequencies[i])
            confocal2.generate_params['rabi_period'] = float(rabi_periods[i])
            confocal2.DEERsim_params['NV2_frequency'] = float(odmr_frequencies[j])
            confocal2.DEERsim_params['NV2_rabiperiod'] = float(rabi_periods[j])
            confocal2.DEERsim_params['tau_start'] = float(max(rabi_periods[i], rabi_periods[j]))
            pulsed_master_logic.set_generation_parameters(confocal2.generate_params)
            confocal2.DEERsim_sweeps = int(5e5)

            print(f"Performing 3pDEER measurement: {deer_label} with {confocal2.DEERsim_sweeps} sweeps")
            confocal2.do_DEERsim(deer_label)
            time.sleep(1)

            tData_deer = netobtain(pulsed_measurement_logic.signal_data[0])
            sigData_deer_bright = netobtain(pulsed_measurement_logic.signal_data[1])
            sigData_deer_dark = netobtain(pulsed_measurement_logic.signal_data[2])
            sigData_deer_bright_err = netobtain(pulsed_measurement_logic.measurement_error[1])
            sigData_deer_dark_err = netobtain(pulsed_measurement_logic.measurement_error[2])
            sigData_deer_subtracted = sigData_deer_bright - sigData_deer_dark
            sigData_deer_subtracted_err = np.sqrt(sigData_deer_bright_err**2 + sigData_deer_dark_err**2)

            # Save and plot raw data
            deer_data_filename = os.path.join(run_folder, f"{deer_label}.txt")
            with open(deer_data_filename, 'w') as f:
                f.write("# tau (s)\tcounts_bright\terr_bright\tcounts_dark\terr_dark\tcounts_subtracted\n")
                for t, sb, sbe, sd, sde, ss in zip(tData_deer, sigData_deer_bright, sigData_deer_bright_err, sigData_deer_dark, sigData_deer_dark_err, sigData_deer_subtracted):
                    f.write(f"{t}\t{sb}\t{sbe}\t{sd}\t{sde}\t{ss}\n")

            plt.figure(figsize=(7, 2.5))
            plt.plot(tData_deer, sigData_deer_bright, 'o-', ms=3, linewidth=0.5, label='Bright State')
            plt.errorbar(tData_deer, sigData_deer_bright, sigData_deer_bright_err, fmt='o', ms=3, linewidth=0.5, ecolor='red')
            plt.plot(tData_deer, sigData_deer_dark, 'o-', ms=3, linewidth=0.5, label='Dark State')
            plt.errorbar(tData_deer, sigData_deer_dark, sigData_deer_dark_err, fmt='o', ms=3, linewidth=0.5, ecolor='blue')
            plt.title(f"DEER Raw Data: {deer_label}")
            plt.xlabel("tau (s)")
            plt.ylabel("counts")
            plt.grid()
            plt.legend()
            deer_raw_fig_path = os.path.join(run_folder, f"{deer_label}_raw.png")
            plt.savefig(deer_raw_fig_path, dpi=150, bbox_inches='tight')
            plt.show()

            # Fit and plot subtracted data for the first DEER measurement
            DEER_rsquared = None # Initialize DEER_rsquared before the try block
            try:
                result_DEER = netobtain(pulsed_measurement_logic.do_fit('ExponentialDecaySine', use_alternative_data=True))
                coupl_freq = result_DEER.params['frequency'].value
                deer_LT = result_DEER.params['decay'].value
                DEER_rsquared = result_DEER.rsquared

                plt.figure(figsize=(7, 2.5))
                plt.plot(tData_deer, sigData_deer_subtracted, 'o-', ms=3, linewidth=0.5, label='Subtracted Data')
                plt.errorbar(tData_deer, sigData_deer_subtracted, sigData_deer_subtracted_err, fmt='o', ms=3, linewidth=0.5, ecolor='red')
                plt.plot(result_DEER.high_res_best_fit[0], result_DEER.high_res_best_fit[1], label='Fit')
                plt.title(f"DEER Subtracted Signal: {deer_label}\nCoupling Freq: {fmt_or_na(coupl_freq, '{:.2E}')} Hz, Decay: {fmt_or_na(deer_LT, '{:.2E}')} s, Rsq: {fmt_or_na(DEER_rsquared, '{:.3f}')}")
                plt.xlabel("tau (s)")
                plt.ylabel("counts")
                plt.grid()
                plt.legend()
                deer_subtracted_fig_path = os.path.join(run_folder, f"{deer_label}_subtracted.png")
                plt.savefig(deer_subtracted_fig_path, dpi=150, bbox_inches='tight')
                plt.show()
            except Exception as e:
                print(f"Failed to fit DEER subtracted data for {deer_label}: {e}")
                plt.figure(figsize=(7, 2.5))
                plt.plot(tData_deer, sigData_deer_subtracted, 'o-', ms=3, linewidth=0.5, label='Subtracted Data')
                plt.errorbar(tData_deer, sigData_deer_subtracted, sigData_deer_subtracted_err, fmt='o', ms=3, linewidth=0.5, ecolor='red')
                plt.title(f"DEER Subtracted Signal: {deer_label} (Fit Failed)")
                plt.xlabel("tau (s)")
                plt.ylabel("counts")
                plt.grid()
                plt.legend()
                deer_subtracted_fig_path = os.path.join(run_folder, f"{deer_label}_subtracted.png")
                plt.savefig(deer_subtracted_fig_path, dpi=150, bbox_inches='tight')
                plt.show()
                DEER_rsquared = -1 # Assign a value < 0.5 to ensure the next block is skipped


            # -----------------------------------------------------------
            # 5b) Same DEER with microwave_2 = 0 (conditional execution)
            # -----------------------------------------------------------
            if selected_routines["DEERsim"]["enabled"] and (DEER_rsquared is not None and DEER_rsquared >= 0.5):
                print(f"DEER {i+1},{j+1} 0 mV started at:", datetime.datetime.now().strftime("%H:%M"))
                deer_label_0mV = f"{currPoi}_{i+1},{j+1}_DEER_0mV"

                # ... (unchanged configuration code for DEER 0mV) ...
                confocal2.generate_params['microwave_frequency'] = float(odmr_frequencies[i])
                confocal2.generate_params['rabi_period'] = float(rabi_periods[i])
                confocal2.DEERsim_params['NV2_frequency'] = float(odmr_frequencies[j])
                confocal2.DEERsim_params['NV2_rabiperiod'] = float(rabi_periods[j])
                confocal2.DEERsim_params['NV2_amplitude'] = float(0.0)
                confocal2.DEERsim_params['tau_start'] = float(max(rabi_periods[i], rabi_periods[j]))
                pulsed_master_logic.set_generation_parameters(confocal2.generate_params)
                confocal2.DEERsim_sweeps = selected_routines["DEERsim"]["sweeps"]

                print(f"Performing 3pDEER measurement: {deer_label_0mV} with {confocal2.DEERsim_sweeps} sweeps")
                confocal2.do_DEERsim(deer_label_0mV)
                time.sleep(1)

                # ... (data retrieval, saving, and plotting for DEER 0mV) ...
                tData_deer_0mV = netobtain(pulsed_measurement_logic.signal_data[0])
                sigData_deer_bright_0mV = netobtain(pulsed_measurement_logic.signal_data[1])
                sigData_deer_dark_0mV = netobtain(pulsed_measurement_logic.signal_data[2])
                sigData_deer_bright_err_0mV = netobtain(pulsed_measurement_logic.measurement_error[1])
                sigData_deer_dark_err_0mV = netobtain(pulsed_measurement_logic.measurement_error[2])
                sigData_deer_subtracted_0mV = sigData_deer_bright_0mV - sigData_deer_dark_0mV
                sigData_deer_subtracted_err_0mV = np.sqrt(sigData_deer_bright_err_0mV**2 + sigData_deer_dark_err_0mV**2)

                deer_data_filename_0mV = os.path.join(run_folder, f"{deer_label_0mV}.txt")
                with open(deer_data_filename_0mV, 'w') as f:
                    f.write("# tau (s)\tcounts_bright\terr_bright\tcounts_dark\terr_dark\tcounts_subtracted\n")
                    for t, sb, sbe, sd, sde, ss in zip(tData_deer_0mV, sigData_deer_bright_0mV, sigData_deer_bright_err_0mV, sigData_deer_dark_0mV, sigData_deer_dark_err_0mV, sigData_deer_subtracted_0mV):
                        f.write(f"{t}\t{sb}\t{sbe}\t{sd}\t{sde}\t{ss}\n")

                # Plot raw data (bright and dark states)
                plt.figure(figsize=(7, 2.5))
                plt.plot(tData_deer_0mV, sigData_deer_bright_0mV, 'o-', ms=3, linewidth=0.5, label='Bright State')
                plt.errorbar(tData_deer_0mV, sigData_deer_bright_0mV, sigData_deer_bright_err_0mV, fmt='o', ms=3, linewidth=0.5, ecolor='red')
                plt.plot(tData_deer_0mV, sigData_deer_dark_0mV, 'o-', ms=3, linewidth=0.5, label='Dark State')
                plt.errorbar(tData_deer_0mV, sigData_deer_dark_0mV, sigData_deer_dark_err_0mV, fmt='o', ms=3, linewidth=0.5, ecolor='blue')
                plt.title(f"DEER Raw Data: {deer_label_0mV}")
                plt.xlabel("tau (s)")
                plt.ylabel("counts")
                plt.grid()
                plt.legend()
                deer_raw_fig_path_0mV = os.path.join(run_folder, f"{deer_label_0mV}_raw.png")
                plt.savefig(deer_raw_fig_path_0mV, dpi=150, bbox_inches='tight')
                plt.show()

                # Fit and plot subtracted data for the DEER 0mV measurement
                try:
                    result_DEER_0mV = netobtain(pulsed_measurement_logic.do_fit('ExponentialDecaySine', use_alternative_data=True))
                    coupl_freq_0mV = result_DEER_0mV.params['frequency'].value
                    deer_LT_0mV = result_DEER_0mV.params['decay'].value
                    DEER_rsquared_0mV = result_DEER_0mV.rsquared

                    plt.figure(figsize=(7, 2.5))
                    plt.plot(tData_deer_0mV, sigData_deer_subtracted_0mV, 'o-', ms=3, linewidth=0.5, label='Subtracted Data')
                    plt.errorbar(tData_deer_0mV, sigData_deer_subtracted_0mV, sigData_deer_subtracted_err_0mV, fmt='o', ms=3, linewidth=0.5, ecolor='red')
                    plt.plot(result_DEER_0mV.high_res_best_fit[0], result_DEER_0mV.high_res_best_fit[1], label='Fit')
                    plt.title(f"DEER Subtracted Signal: {deer_label_0mV}\nCoupling Freq: {fmt_or_na(coupl_freq_0mV, '{:.2E}')} Hz, Decay: {fmt_or_na(deer_LT_0mV, '{:.2E}')} s, Rsq: {fmt_or_na(DEER_rsquared_0mV, '{:.3f}')}")
                    plt.xlabel("tau (s)")
                    plt.ylabel("counts")
                    plt.grid()
                    plt.legend()
                    deer_subtracted_fig_path_0mV = os.path.join(run_folder, f"{deer_label_0mV}_subtracted.png")
                    plt.savefig(deer_subtracted_fig_path_0mV, dpi=150, bbox_inches='tight')
                    plt.show()
                except Exception as e:
                    print(f"Failed to fit DEER 0mV subtracted data for {deer_label_0mV}: {e}")
                    plt.figure(figsize=(7, 2.5))
                    plt.plot(tData_deer_0mV, sigData_deer_subtracted_0mV, 'o-', ms=3, linewidth=0.5, label='Subtracted Data')
                    plt.errorbar(tData_deer_0mV, sigData_deer_subtracted_0mV, sigData_deer_subtracted_err_0mV, fmt='o', ms=3, linewidth=0.5, ecolor='red')
                    plt.title(f"DEER Subtracted Signal: {deer_label_0mV} (Fit Failed)")
                    plt.xlabel("tau (s)")
                    plt.ylabel("counts")
                    plt.grid()
                    plt.legend()
                    deer_subtracted_fig_path_0mV = os.path.join(run_folder, f"{deer_label_0mV}_subtracted.png")
                    plt.savefig(deer_subtracted_fig_path_0mV, dpi=150, bbox_inches='tight')
                    plt.show()
            else:
                print("Skipping DEER 0mV measurement due to low R-squared from the main DEER fit.")
```
